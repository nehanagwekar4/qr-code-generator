# QR CODE With Personalized LOGO



```python
!pip install qrcode[pil]

import qrcode
from PIL import Image


# User Input
data = input("Enter text or URL: ")

#Image for Logo
from google.colab import files
uploaded = files.upload()

logo_path = list(uploaded.keys())[0]

print("Uploaded:", logo_path)

# Create QR Code
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_H,   # High error correction
    box_size=10,
    border=4
)

qr.add_data(data)
qr.make(fit=True)

img = qr.make_image(
    fill_color="pink",
    back_color="white"
).convert("RGB")


# Load Logo
logo = Image.open(logo_path).convert("RGBA")

# Resize logo
qr_width, qr_height = img.size

logo_size = qr_width // 5      # Logo = 20% of QR width

logo = logo.resize((logo_size, logo_size))

# Position logo at center
x = (qr_width - logo_size) // 2
y = (qr_height - logo_size) // 2

# Paste logo
img.paste(logo, (x, y), mask=logo if logo.mode == "RGBA" else None)

# Save
img.save("QR_with_logo.png")

print("QR Code Generated Successfully!")

img
```
